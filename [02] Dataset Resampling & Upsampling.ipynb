### The purpose of this code is to identify which gesture need to increase (augmentation) and which to decrease (resampling) to reduce biases in dataset.

```python
import numpy as np
import os
import random
import shutil
```

```python
video_directory = 'TRAIN_4'

sum = 0

gesture_folder = np.array(os.listdir(video_directory))
for gestures in gesture_folder:
    gesture = []

    for fname in os.listdir(os.path.join(video_directory, gestures)):
        path = os.path.join(video_directory, gestures, fname)
        if os.path.isdir(path):
            gesture.append(fname)

    sum += len(gesture) 
    if len(gesture) != 0:
    # if len(gesture) > 40:
        print(gestures, end =" : ")        
        print(len(gesture))
        
        

print("Total gestures: ", len(gesture_folder), "; Total videos: ", sum)
```

```python
# Parameters
max_videos_per_gesture = 40  # Maximum videos per gesture
min_videos_per_gesture = 31  # Average videos per gesture
video_directory = 'TRAIN_3'
output_directory = "TRAIN_4"

# Create output directory structure
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Loop through gestures
gesture_folder = np.array(os.listdir(video_directory))
for gestures in gesture_folder:
    gesture = []

    for fname in os.listdir(os.path.join(video_directory, gestures)):
        path = os.path.join(video_directory, gestures, fname)
        if os.path.isdir(path):
            gesture.append(fname)

    if len(gesture) <= max_videos_per_gesture:
        for video in gesture:
            src = os.path.join(video_directory, gestures, video)
            dest = os.path.join(output_directory, gestures, video)
            os.makedirs(dest, exist_ok=True)
            for npy in os.listdir(src):
                src_file = os.path.join(src, npy)
                dest_file = os.path.join(dest, npy)
                shutil.copy(src_file, dest_file)

    if len(gesture) < min_videos_per_gesture:
        upsample_count = min_videos_per_gesture - len(gesture)
        selected_videos = random.choices(gesture, k=upsample_count)

        # print(upsample_count)
        # print(selected_videos)

        # Copy videos to output directory
        copy = [int(gesture[-1]) + i + 1 for i in range(upsample_count)]
        # print("copy", copy)

        for video, copy in zip(selected_videos, copy):
            src = os.path.join(video_directory, gestures, video)
            dest = os.path.join(output_directory, gestures, str(copy))
            os.makedirs(dest, exist_ok=True)
            for npy in os.listdir(src):
                src_file = os.path.join(src, npy)
                dest_file = os.path.join(dest, npy)
                shutil.copy(src_file, dest_file)

        print(f"{gestures}: Upsampling using {selected_videos}")

```

```python
# Parameters
max_videos_per_gesture = 40
video_directory = 'TRAIN_3'
output_directory = "TRAIN_4"

# Create output directory structure
# if not os.path.exists(output_directory):
#     os.makedirs(output_directory)

# Loop through gestures
gesture_folder = np.array(os.listdir(video_directory))
for gestures in gesture_folder:
    gesture = []

    for fname in os.listdir(os.path.join(video_directory, gestures)):
        path = os.path.join(video_directory, gestures, fname)
        if os.path.isdir(path):
            gesture.append(fname)

    if len(gesture) > max_videos_per_gesture:
        resample_count = max_videos_per_gesture
        selected_videos = random.sample(gesture, k=resample_count)

        # print(upsample_count)
        # print(selected_videos)

        # Copy videos to output directory
        copy = [i + 1 for i in range(resample_count)]
        # print("copy", copy)

        for video, copy in zip(selected_videos, copy):
            src = os.path.join(video_directory, gestures, video)
            dest = os.path.join(output_directory, gestures, str(copy))
            os.makedirs(dest, exist_ok=True)
            for npy in os.listdir(src):
                src_file = os.path.join(src, npy)
                dest_file = os.path.join(dest, npy)
                shutil.copy(src_file, dest_file)

        print(f"{gestures}: Resampling using {selected_videos} only")
```
