```python
# Importing the required libraries
import os
import shutil
```

```python
%%script false --no-raise-error
def copy_folders(src, dest):

    for root, dirs, files in os.walk(src):
        if "interpolated_landmarks" in dirs:
            interpolated_path = os.path.join(root, "interpolated_landmarks")

            relative_path = os.path.relpath(root, src)
            
            target_path = os.path.join(dest, relative_path)

            os.makedirs(target_path, exist_ok=True)

            for file_name in os.listdir(interpolated_path):
                src_file = os.path.join(interpolated_path, file_name)
                dest_file = os.path.join(target_path, file_name)
                
                shutil.copy(src_file, dest_file)
                print(f"Copied {src_file} to {dest_file}")

src = "DATASET"
dest = "TRAIN"                

copy_folders(src, dest)
```

```python
# %%script false --no-raise-error

def copy_half_folders(src, dest):
    main_folders = [d for d in os.listdir(src) if os.path.isdir(os.path.join(src, d))]

    for folder in main_folders:
        source_path = os.path.join(src, folder)
        target_path = os.path.join(dest, folder)
        
        os.makedirs(target_path, exist_ok=True)
        
        subfolders = sorted([d for d in os.listdir(source_path) if os.path.isdir(os.path.join(source_path, d))])
        
        for subfolder in subfolders[:int(len(subfolders)/2)]:
            source_subfolder_path = os.path.join(source_path, subfolder)
            destination_subfolder_path = os.path.join(target_path, subfolder)
            shutil.copytree(source_subfolder_path, destination_subfolder_path)
            print(f"Copied {source_subfolder_path} to {destination_subfolder_path}")

src = 'TRAIN'
dest = 'TRAIN_3'

copy_half_folders(src, dest)
```
