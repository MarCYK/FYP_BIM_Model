```python
import cv2
import os
import mediapipe as mp

# Initialize MediaPipe and Drawing Utils
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils

filename = 'withhand.mp4'
frames_per_second = 24.0
res = '720p'

# Change resolution
def change_res(cap, width, height):
    cap.set(3, width)
    cap.set(4, height)

# Standard dimensions for resolutions
STD_DIMENSIONS = {
    "480p": (640, 480),
    "720p": (1280, 720),
    "1080p": (1920, 1080),
    "4k": (3840, 2160),
}

# Get dimensions based on resolution
def get_dims(cap, res='1080p'):
    width, height = STD_DIMENSIONS["480p"]
    if res in STD_DIMENSIONS:
        width, height = STD_DIMENSIONS[res]
    change_res(cap, width, height)
    return width, height

# Video type supported
VIDEO_TYPE = {
    'avi': cv2.VideoWriter_fourcc(*'XVID'),
    'mp4': cv2.VideoWriter_fourcc(*'XVID'),
}

def get_video_type(filename):
    _, ext = os.path.splitext(filename)
    if ext in VIDEO_TYPE:
        return VIDEO_TYPE[ext]
    return VIDEO_TYPE['avi']

cap = cv2.VideoCapture(0)
dims = get_dims(cap, res)
out = cv2.VideoWriter(filename, get_video_type(filename), frames_per_second, dims)

with mp_hands.Hands(min_detection_confidence=0.8, min_tracking_confidence=0.5) as hands:
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Save the unmodified frame to the output video
        out.write(frame)

        # Process the frame for hand tracking
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image = cv2.flip(image, 1)
        image.flags.writeable = False
        result = hands.process(image)
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        try:
            if result.multi_hand_landmarks:
                for num, hand in enumerate(result.multi_hand_landmarks):
                    mp_drawing.draw_landmarks(
                        image, hand, mp_hands.HAND_CONNECTIONS,
                        mp_drawing.DrawingSpec(color=(15, 55, 250), thickness=2, circle_radius=6),
                        mp_drawing.DrawingSpec(color=(15, 55, 250), thickness=2, circle_radius=4),
                    )
        except Exception as e:
            print(e)

        # Show the frame with hand landmarks
        cv2.imshow('hand tracking', image)

        # Break loop on 'q' key press
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()
```
